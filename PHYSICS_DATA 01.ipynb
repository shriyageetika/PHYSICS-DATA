**TIME**
: In many datasets, 'time' represents an independent variable that tracks the duration or point in an event or process. It could be measured in seconds, minutes, hours, days, or any other time unit.

**FLUX INTEGRAL**: A flux integral calculates the total flow of a quantity (like heat, mass, or fluid) across a surface or boundary over a certain period.
It represents the total amount of the quantity passing through a defined area.

**INLET T**: Inlet temperature (Tin) usually refers to the temperature of a substance (like a fluid or gas) as it enters a system or a control volume.

**MAX T**: Maximum temperature (Tmax) is the highest temperature observed within a system or during a specific time frame.
This would be the highest value recorded in a temperature-related column in our dataset.

**OUTLET T**: Outlet temperature (Tout) refers to the temperature of a substance as it exits a system or a control volume.
Similar to inlet temperature, this is the temperature recorded at the exit point.

```python
import matplotlib.pyplot as plt

# Assuming 'df' has 'time' as index and other columns: 'flux_integral', 'inlet_T', 'max_T', 'outlet_T'
# Plot with 'max_T' on the x-axis
plt.figure(figsize=(7,4))
plt.plot(df['time'],df['max_T'])
plt.title('TIME DATA max_T')
plt.xlabel('time')
plt.ylabel('max_T')
plt.show()

# Plot with 'inlet_T' on the x-axis
plt.figure(figsize=(7,4))
plt.plot(df['time'], df['inlet_T'])  # Use data.index to access the 'time' values
plt.title('TIME DATA inlet_T')
plt.xlabel('time')
plt.ylabel('inlet_T')
plt.show()
```

Subsetting Time Using Pandas Dataframes

```python
# Assuming 'df' has 'time' as index and 'flux_integral' as a column
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.plot(df['time'], df['flux_integral'])  # Plot 'flux_integral' against 'time'
plt.title('Flux Integral over Time')
plt.xlabel('Time')
plt.ylabel('Flux Integral')
plt.grid(True)  # Add a grid for better readability
plt.show()
```

```python
df_time = df.set_index('time')
df_time.head()
ax = df_time.plot(fontsize=10,figsize=(6,4),title='DATA TIME')
# Plotting both inlet_T and outlet_T
ax.plot(df['time'], df['inlet_T'], linestyle='-', alpha=0.7)
ax.plot(df['time'], df['outlet_T'], linestyle='--', alpha=0.7)
ax.set_xlabel('time')
ax.set_ylabel('Temperature')
ax.axhspan(450, 400, color='black', alpha=0.3)
plt.legend()

# Set y-axis range
ax.set_ylim(0, 1000)

plt.show()
```

#**TIME RESAMPLING**

```python
df_time = df.set_index('time')
df_time.index = pd.to_datetime(df_time.index) # Convert index to DatetimeIndex
df_time.resample(rule='H').mean(numeric_only=True)[:5]
```

```python
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace= True)
```
